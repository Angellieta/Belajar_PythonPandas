```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size =(n_rows, n_cols)),
                  columns=cols)
df
#ini prosesnya sama, cuma berbeda di jumlah baris dan angka random yang munculnya ( yang muncul 1 2 3 4 5 )#
```

```python
df[(df['A'] == 1) | (df['A'] == 3)]
# A nya keluar hanya bilangan 1 dan 3 #
#ini adalah operator logika or atau ( | ) #
```

```python
df[df['A'].isin([1, 3])]
#operator logika or dapet digantikan dengan memanfaatkan fungsi .isin()#
```

```python
df[~df['A'].isin([1, 3])]
#pemilihan angka yang bukan angka 1 dan 3#
#operator negasi (~) untuk perintah not#
```

```python
data = {'col1':['1','2','3','teks'],
       'col2':['1','2','3','4']}
df = pd.DataFrame(data) #mau buat data frame yang didapet dari data diatas#
df
#mempersiapkan data frame untuk kolom 1 dan kolom 2#
```

```python
df.dtypes
#ini gunanya meninjau tipe datanya#
#kenapa hasilnya object semua?, karena dalam data semuanya string yang memakai tanda petik"
#data string adalah data object#
```

```python
df_x = df.astype({'col2':'float'})
df_x
#ini adalah konversi tipe data dari string ke int ataupun ke float satu kolom menggunakan fungsi .astype()#
```

```python
df_x.dtypes
```

```python
df.apply(pd.to_numeric, errors='coerce')
#fungsi to_numeric() gunanya mengkonversi tipe data ke tipe data numerik untuk semua kolom#
#fungsi errornya coerce adalah jika konversi data di ubah ke tipe data numerik, data yang tidak dapat dikonversikan maka data itu akan digantikan dengan kata NaN atau disebut missing value#
```

```python
n_rows = 5
n_cols = 2
cols = ['bil_pecahan','bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size =(n_rows, n_cols)),
                  columns=cols)
#buat dataframe yang nama kolomnya bilangan pecahan dan bilangan bulat, dengan isi nilai random dari range 1 sampe 20#
df['bil_pecahan'] = df['bil_pecahan'].astype('float') #lalu dikonversikan isi dari bilangan pecahan menjadi float#

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
#untuk membuat kolom index dan frekuensi#
df = df.reset_index()

df['teks'] = list('ABCDE')
#untuk membuat kolom teks#
df
```

```python
df.dtypes
#memverifikasi tipe data#
```

```python
df.select_dtypes(include = 'number')
#fungsi diatas akan memilih tipe data yang numerik saja#
```

```python
df.select_dtypes(include = 'float')
#fungsi diatas akan memilih tipe data yang float saja#
```

```python
df.select_dtypes(include = 'int') #tinggal ubah parameternya saja yang berwarna merah#
#fungsi diatas akan memilih tipe data yang int saja#
```

```python
df.select_dtypes(include = 'object')
#fungsi diatas akan memilih tipe data String atau object saja#
```

```python
df.select_dtypes(include = 'datetime')
#fungsi diatas akan memilih tipe data datetime saja#
```

```python
df.select_dtypes(include = ['number','object'])
#fungsi ini untuk memilih tipe data yang dikombinasi#
```
